El evento más frecuente es first_visit, seguido por acciones relacionadas con la exploración y compra de productos. Sin embargo, para construir correctamente el funnel será necesario analizar usuarios únicos por etapa, ya que un mismo usuario puede generar múltiples eventos durante su recorrido.

```python
# PARTE 1: Totales del funnel
# ======================

query_totals = '''

SELECT
    nombre_evento,
    COUNT(DISTINCT id_usuario) AS usuarios,
CASE
    WHEN nombre_evento = 'first_visit' THEN 1
    WHEN nombre_evento = 'select_item' THEN 2
    WHEN nombre_evento = 'add_to_cart' THEN 3
    WHEN nombre_evento = 'begin_checkout' THEN 4
    WHEN nombre_evento = 'add_payment_info' THEN 5
    WHEN nombre_evento = 'purchase' THEN 6
END AS orden
FROM events
GROUP BY nombre_evento, orden
ORDER BY orden ASC;


'''

totals = pd.read_sql(query_totals, con=engine)
totals

```

- Se construyó el funnel de conversión utilizando usuarios únicos por evento. La etapa con mayor alcance fue ***first_visit***, con 7,796 usuarios, mientras que 6,240 usuarios completaron una compra. En general, se observa una disminución gradual de usuarios a medida que avanzan por las distintas etapas del proceso de compra, lo que indica que existen pérdidas de usuarios a lo largo del recorrido hasta la conversión final.
- Se observa una conversión superior al 100% entre select_item y add_to_cart. Esto sugiere que algunos usuarios llegaron directamente al carrito sin registrar previamente el evento select_item, una situación común cuando existen rutas alternativas dentro de la aplicación o eventos faltantes en la instrumentación.

```python

# PARTE 2: Conversiones
# ======================

query_conversion = '''
WITH cte_first_visit as (
    SELECT DISTINCT id_usuario
    FROM events
    WHERE nombre_evento = 'first_visit'
),
cte_select_item as (
    SELECT DISTINCT id_usuario
    FROM events
    WHERE nombre_evento = 'select_item'
),
cte_add_to_cart as (
    SELECT DISTINCT id_usuario
    FROM events
    WHERE nombre_evento = 'add_to_cart'
),
cte_begin_checkout as (
    SELECT DISTINCT id_usuario
    FROM events
    WHERE nombre_evento = 'begin_checkout'
),
cte_add_payment_info as (
    SELECT DISTINCT id_usuario
    FROM events
    WHERE nombre_evento = 'add_payment_info'
),
cte_purchase as (
    SELECT DISTINCT id_usuario
    FROM events
    WHERE nombre_evento = 'purchase'
)
SELECT
    (SELECT COUNT(*) FROM cte_first_visit) AS first_visit_users,

    (SELECT COUNT(*) FROM cte_select_item) AS select_item_users,
    (SELECT COUNT(*) FROM cte_add_to_cart) AS add_to_cart_users,
    (SELECT COUNT(*) FROM cte_begin_checkout) AS begin_checkout_users,
    (SELECT COUNT(*) FROM cte_add_payment_info) AS add_payment_info_users,
    (SELECT COUNT(*) FROM cte_purchase) AS purchase_users,

    ROUND((SELECT COUNT(*) FROM cte_select_item) * 100.00 / NULLIF((SELECT COUNT(*) FROM cte_first_visit), 0), 2) AS conversion_select_from_first,
    ROUND((SELECT COUNT(*) FROM cte_add_to_cart) * 100.00 / NULLIF((SELECT COUNT(*) FROM cte_select_item), 0), 2) AS conversion_cart_from_select,
    ROUND((SELECT COUNT(*) FROM cte_begin_checkout) * 100.00 / NULLIF((SELECT COUNT(*) FROM cte_add_to_cart), 0), 2) AS conversion_checkout_from_cart,
    ROUND((SELECT COUNT(*) FROM cte_add_payment_info) * 100.00 / NULLIF((SELECT COUNT(*) FROM cte_begin_checkout), 0), 2) AS conversion_payment_from_checkout,
    ROUND((SELECT COUNT(*) FROM cte_purchase) * 100.00 / NULLIF((SELECT COUNT(*) FROM cte_add_payment_info), 0), 2) AS conversion_purchase_from_payment,
    ROUND((SELECT COUNT(*) FROM cte_purchase) * 100.00 / NULLIF((SELECT COUNT(*) FROM cte_first_visit), 0), 2) AS conversion_purchase_from_first;
    
    

'''


conversion = pd.read_sql(query_conversion, con=engine)
conversion
```

El análisis del funnel muestra una conversión final de 80.04% desde first_visit hasta purchase. La mayor pérdida de usuarios ocurre entre las etapas begin_checkout y add_payment_info, donde la conversión es de 86.71%, lo que sugiere posibles fricciones durante el proceso de pago. Asimismo, se observa una conversión superior al 100% entre select_item y add_to_cart, lo que podría indicar diferencias en el registro de eventos o recorridos alternativos de los usuarios dentro de la plataforma. Aunque la conversión final es elevada (80.04%), la mayor oportunidad de optimización se encuentra entre las etapas de exploración y selección de productos, donde ocurre la mayor pérdida de usuarios.

## 🔹 Paso 4: Evaluar si los usuarios regresan (retención por cohortes)

**🎯 Objetivo:** Analizar la retención de usuarios para entender si regresan después de registrarse.

**Tablas**

- `users` 
- `user_activity` 

---
1. Se identifica la cohorte de cada usuario según el **mes de registro**.


2. Se calcula la retención semanal: cuántos usuarios **se mantienen activos** en cada semana desde su registro.
   - `retenido_w1`: usuarios activos en la semana 1  
   - `retenido_w2`: usuarios activos en la semana 2  
   - `retenido_w3`: usuarios activos en la semana 3  


3. Se calcula el porcentaje de retención para cada semana, dividiendo los usuarios retenidos entre los clientes iniciales de la cohorte:  
   - `semana_1`: porcentaje de usuarios retenidos en la semana 1  
   - `semana_2`: porcentaje de usuarios retenidos en la semana 2  
   - `semana_3`: porcentaje de usuarios retenidos en la semana 3  

Se revisa que la columna de fecha esté en formato correcto (`DATE`).  
Se realiza la conversión usando: `CAST(fecha_registro AS DATE)`

```python
# Explorar tabla users
# =========================
query_users = '''
SELECT *
FROM users;
'''
users = pd.read_sql(query_users, con=engine)
users.head(3)
```

```python
# Explorar tabla user_activity
# =========================
query_user_activity = '''
SELECT *
FROM user_activity;
'''
user_activity = pd.read_sql(query_user_activity, con=engine)
user_activity.head(3)
```

```python

# Retención por cohortes
# ======================

query_cohort_retention_final = '''
WITH creacion_cohortes_mensuales AS (    
    SELECT
    u.id_usuario,
    CAST(u.fecha_registro AS DATE) AS fecha_registro_date,
    DATE_TRUNC('month', CAST(u.fecha_registro AS DATE)) AS cohortes_mensuales,
    CAST(ua.fecha_actividad AS DATE) AS fecha_actividad_date,
    ua.dias_despues_registro,
    ua.activo
    FROM users u
    LEFT JOIN user_activity ua ON u.id_usuario = ua.id_usuario
    WHERE ua.activo = 1
    ),


retencion_semanal AS (
    SELECT
        cohortes_mensuales,
        COUNT(DISTINCT id_usuario) AS usuarios_iniciales,
        COUNT(DISTINCT CASE WHEN dias_despues_registro >= 1 AND dias_despues_registro <= 7 THEN id_usuario END) AS retenido_w1,
        COUNT(DISTINCT CASE WHEN dias_despues_registro >= 8 AND dias_despues_registro <= 14 THEN id_usuario END) AS retenido_w2,
        COUNT(DISTINCT CASE WHEN dias_despues_registro >= 15 AND dias_despues_registro <= 21 THEN id_usuario END) AS retenido_w3
        FROM creacion_cohortes_mensuales
        GROUP BY cohortes_mensuales
)

SELECT 
    TO_CHAR(cohortes_mensuales, 'YYYY-MM') AS cohorte,
    usuarios_iniciales,
    retenido_w1,
    ROUND(retenido_w1::numeric/NULLIF(usuarios_iniciales, 0)*100, 2) AS semana_1_pct,
    retenido_w2,
    ROUND(retenido_w2::numeric/NULLIF(usuarios_iniciales, 0)*100, 2) AS semana_2_pct,
    retenido_w3,
    ROUND(retenido_w3::numeric/NULLIF(usuarios_iniciales, 0)*100, 2) AS semana_3_pct
FROM retencion_semanal;


'''

# Ejecutar la consulta
cohorte_final = pd.read_sql(query_cohort_retention_final, con=engine)
cohorte_final
```

<div class="alert alert-block alert-danger"> <b>Comentario del revisor. (Iteración 1)</b> <a class="tocSkip"></a><br><br>

Esta sección necesita corrección antes de poder considerarse validada. La idea general de cohortes está bien orientada porque partes de fecha_registro para crear cohortes_mensuales, pero hay dos problemas técnicos importantes en la lógica de retención.

Primero, en la CTE inicial filtras con WHERE ua.activo = 1. Eso hace que usuarios_iniciales se calcule solo sobre usuarios que tuvieron actividad activa, no sobre todos los usuarios registrados en la cohorte. En retención, el denominador debe representar la base inicial de usuarios de cada cohorte; si excluimos a los inactivos desde el inicio, la retención queda inflada.

Segundo, estás calculando retenido_w1, retenido_w2 y retenido_w3 con rangos exactos de días (1 a 7, 8 a 14, 15 a 21). La rúbrica espera una lógica semanal acumulada, usando una semana derivada de dias_despues_registro / 7 y condiciones tipo semana mayor o igual a 1, 2 y 3. Si usas semanas exactas, estás midiendo actividad puntual en cada rango, no permanencia o regreso acumulado.

Para corregirlo, revisa la base de usuarios iniciales desde users, calcula la semana a partir de dias_despues_registro y luego construye la retención acumulada sobre esa base. Este ajuste es clave porque la retención es una métrica sensible: un denominador incorrecto puede llevar al negocio a pensar que los usuarios regresan más de lo que realmente ocurre.

---
Hipótesis estadística
   - **H₀ (Hipótesis nula):** La proporción de conversión del grupo control es igual a la proporción de conversión del grupo tratamiento.
   - **H₁ (Hipótesis alternativa):** La proporción de conversión del grupo control es diferente a la proporción de conversión del grupo tratamiento.
   
**Test estadístico:** Prueba Z para comparación de dos proporciones independientes  
**Nivel de significancia alpha:** alpha = 0.05

```python
observaciones = [total_usuarios['control'], total_usuarios['tratamiento']]
observaciones
```

```python
z_stat, p_value = proportions_ztest(conversiones_grupo, observaciones)

print(f'Estadistico ztest: {z_stat}')
print(f'Valor p: {p_value}')
```

```python
alpha = 0.05

if p_value < alpha:
    print('Rechazamos la hipótesis nula: hay evidencia de una diferencia.')
else:
    print('No rechazamos la hipótesis nula: no hay evidencia suficiente de una diferencia.')
```

El grupo tratamiento presentó una tasa de conversión de 16.29%, ligeramente superior a la del grupo control (15.69%). Para determinar si esta diferencia es estadísticamente significativa, se aplicó una prueba Z para comparación de dos proporciones independientes con un nivel de significancia de 0.05. El resultado arrojó un estadístico Z de -0.8133 y un valor p de 0.4161. Debido a que el valor p es mayor que α = 0.05, no se rechaza la hipótesis nula. Por lo tanto, no existe evidencia estadística suficiente para afirmar que la modificación de la interfaz del checkout haya generado un impacto significativo en la tasa de conversión.

No se recomienda implementar el cambio únicamente por su efecto en la conversión, ya que la diferencia observada podría deberse al azar. Se sugiere continuar evaluando otras métricas o realizar nuevas pruebas para confirmar el impacto del cambio.
